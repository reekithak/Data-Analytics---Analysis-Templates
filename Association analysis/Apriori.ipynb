```python
import pandas as pd
import numpy as np 
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
sheet = pd.read_excel('association.xls',sheet_name='Chandigarh')
```

```python
def work_load(sheet):
    work = sheet[['Invoice #','SKU','PCS']]
    uniq_sku = work['SKU'].unique()
    tmp = work.groupby(['Invoice #','SKU'])['PCS'].sum().unstack().reset_index().fillna(0).set_index('Invoice #')
    new_set = tmp.applymap(encode_units)
    frequent_itemsets = apriori(new_set, min_support=0.05, use_colnames=True)
    return frequent_itemsets
freq = work_load(sheet)   
freq['Length'] = freq['itemsets'].apply(lambda x :len(x)) 
```
